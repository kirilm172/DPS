### 1. Merges the training and the test sets to create one data set.

```python
features = pd.read_csv('UCI HAR Dataset/features.txt',
                       header=None, delim_whitespace=True)
features.columns = ['id', 'feature label']
features = features.set_index('id')
features
```

```python
activity_labels = pd.read_csv(
    'UCI HAR Dataset/activity_labels.txt', header=None, delim_whitespace=True)
activity_labels.columns = ['id', 'activity label']
activity_labels = activity_labels.set_index('id')
activity_labels
```

```python

test_x = pd.read_csv('UCI HAR Dataset/test/X_test.txt',
                     header=None, delim_whitespace=True)
test_y = pd.read_csv('UCI HAR Dataset/test/Y_test.txt',
                     header=None, delim_whitespace=True)

train_x = pd.read_csv('UCI HAR Dataset/train/X_train.txt',
                      header=None, delim_whitespace=True)
train_y = pd.read_csv('UCI HAR Dataset/train/Y_train.txt',
                      header=None, delim_whitespace=True)

test_subject = pd.read_csv('UCI HAR Dataset/test/subject_test.txt',
                           header=None, delim_whitespace=True)
train_subject = pd.read_csv('UCI HAR Dataset/train/subject_train.txt',
                            header=None, delim_whitespace=True)


df = pd.concat([test_x, train_x])
y = pd.concat([test_y, train_y])
subjects = pd.concat([test_subject, train_subject])

df.columns = list(features['feature label'])
y.columns = ['label']
subjects.columns = ['subject']

df['activity'] = y['label'].map(
    lambda x: activity_labels.loc[x]['activity label'])
df['subject'] = subjects['subject']
df = df.dropna(how='any')
df = df.drop_duplicates()
df = df.reset_index()
del df['index']

df
```
